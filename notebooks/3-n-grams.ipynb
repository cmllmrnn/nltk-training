```
from nltk import ngrams

# load our sample string/input
sentence = 'The Solar System is the gravitationally bound system of the planets and the Sun plus other objects that orbit it, either directly or indirectly. Of the objects that orbit the Sun directly, the largest eight are the planets, with the remainder being smaller objects, such as dwarf planets and small Solar System bodies. Of the objects that orbit the Sun indirectly, the moons, two are larger than the smallest planet, Mercury'

# unigram
n = 1 

# apply the ngrams class of nltk
unigrams = ngrams(sentence, n)

# print the output of the unigram
print('Unigrams: {}'.format(unigrams))
```

```
for gram in unigrams:
  print(gram, end=' ')
```

```
# bigram
n = 2
bigrams = ngrams(sentence, n)

for gram in bigrams:
  print(gram, end=' ')
```

```
# trigram
n = 3
trigrams = ngrams(sentence, n)

for gram in trigrams:
  print(gram, end=' ')
```

```
# split the sentence
words = sentence.split()

print(words)
```

```
# loop three iterations
for count in range(3):
  n = count+1
  
  #save our current ngrams
  current_grams = ngrams(words, n)
  
  if n==1:
    print('Unigrams:')
  elif n==2:
    print('\nBigrams:')
  elif n==3:
    print('\nTrigrams:')
    
  # print the n-grams of current n-gram
  for gram in current_grams:
    print(gram, end=' ')
```

```
with open('reviews.txt', 'r') as file:
  corpus = file.read().split()
```

```
# save the unigrams here
unigram_list = []

# unigrams
n = 1

unigrams = ngrams(corpus, n)

for gram in unigrams:
  unigram_list.append(gram)
```

```
print(unigram_list[:20])
```

```
from collections import defaultdict

dictionary = defaultdict(list)

# make a temporary value for the previous word
previous_word = ''

# loop for appending values to the dictionary
for current_word in unigram_list:
  if previous_word != '':
    dictionary[previous_word].append(current_word[0])
  previous_word = current_word[0]
  
dictionary
```

```
# define a function for building conditional probabilities
def compute_conditional_probabilities(corpus):
  for key in dictionary.keys():
    # getting next word from dictionary
    next_word = dictionary[key]
    
    # remove the duplicate words and save it
    unique_words = set(next_word)
    
    num_words = len(next_word)
    
    # placeholder dictionary for the loop
    probabilities_given_key = {}
    
    # loop for saving value of current key
    for unique_word in unique_words:
      # save the value of the current word
      probabilities_given_key[unique_word] = float(next_word.count(unique_word)) / num_words
  return probabilities_given_key
```

```
# define a function for predicting the probability of the next word
def next_word_predictor(conditional_probabilities, current_word, next_word):
  # if current word is existing, check probability for next word
  if current_word in conditional_probabilities:
    # if conditional probability of both words exists, return value
    if next_word in conditional_probabilities[current_word]:
      return conditional_probabilities[current_word][next_word]
  return 0.0
```

```
conditonal_probabilities = compute_conditional_probabilities(corpus)
```
